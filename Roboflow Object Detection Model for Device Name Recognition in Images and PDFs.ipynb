#6)Reading from pdf file

Creatin custom torch model 



```python
model = torch.hub.load('.', 'custom', path='/content/yolov5/runs/train/exp/weights/best.pt', source='local') 
```

```python
images=convert_from_bytes(open(r'/content/Label.pdf','rb').read())
```

```python
for i,image in enumerate(images):
    f_name="image"+str(i)+".png"

    image.save(f_name,'png')
    img=cv2.imread(f_name)
    img = cv2.resize(img, (960, 540))   

    reader = easyocr.Reader(['en'])
    output = reader.readtext(img)
    img = f_name
    # Inference
    results = model(img)
    # Results, change the flowing to: results.show()
    temp=results.pandas().xyxy[0]
    x=temp['class'] # or .show(), .save(), .crop(), .pandas(), etc
    res=""
    for i in x:
      res=res+str(i)
    text=""
    for i in output:
        text=text+i[1]+"  "
    d_name=re.search("Device Name: ([A-Za-z\W]{2,9}[A-Za-z]*)",text)[1]
    ref_no=re.search("REF ([A-Za-z\d]{2,})",text)[1]
    qty_no=qty_no=re.search("Qty: (\d+|\w)",text)
    if(qty_no==None):
        qty_no=qty_no
    else:
        qty_no=int(qty_no[1])
    lot_no=re.search("LOT: (\d+)",text)[1]
    ws.append([d_name,ref_no,lot_no,qty_no,res])
```
